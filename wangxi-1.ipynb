```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('avgHgt.csv')
df=pd.DataFrame(data)
df.columns = ['age','CHeight','JHeight']
plt.title('height figure')
plt.xlabel('age')
plt.ylabel('height')
plt.plot(df['age'],df['CHeight'],label='China')
plt.plot(df['age'],df['JHeight'],label='Japan')
plt.legend()

data.to_csv('avgHgt.txt',index=0)
```
